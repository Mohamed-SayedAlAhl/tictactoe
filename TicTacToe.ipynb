```python
board = [[' ', ' ', ' '], 
         [' ', ' ', ' '], 
         [' ', ' ', ' ']]

def print_board():
    print("\n")
    print(f"{board[0][0]}|{board[0][1]}|{board[0][2]}")
    print("-+-+-")
    print(f"{board[1][0]}|{board[1][1]}|{board[1][2]}")
    print("-+-+-")
    print(f"{board[2][0]}|{board[2][1]}|{board[2][2]}")
    print("\n")
```

```python
def get_move(player):

    while True:
        print(f"\n{player}'s turn.")
        row = input("Enter row (0, 1, or 2): ")
        col = input("Enter column (0, 1, or 2): ")
        
        try:
            row = int(row)
            col = int(col)
        except ValueError:
            print("Please enter an integer between 0 and 2.")
            continue
        
        if row < 0 or row > 2 or col < 0 or col > 2:
            print("Index out of range. Please try again.")
            continue
            
        if board[row][col] != ' ':
            print("That space is taken!")
            continue
        
        break
        
    board[row][col] = player
    print_board()
```

```python
def check_win(player):
    # check rows
    for row in board:
        if all(spot == player for spot in row):
            print(f"{player} wins!")
            return True
    # check columns
    for i in range(3):
        if (board[0][i] == player and board[1][i] == player and board[2][i] == player):
            print(f"{player} wins!")
            return True
        # check diagonals
        if (board[0][0] == player and board[1][1] == player and board[2][2] == player):
            print(f"{player} wins!")
            return True
        
        if (board[0][2] == player and board[1][1] == player and board[2][0] == player):
            print(f"{player} wins!")
            return True
    return False
```

```python
def check_tie():
  
    for row in board:
        if ' ' in row:
            return False
  
    print("It's a tie!")
    return True
```

```python
def play_game():

    players = ['X', 'O']
    current_player = players[0]

    while True:

        get_move(current_player)

        if check_win(current_player):
            break
    
        if check_tie():
            break
    
        current_player = players[0] if current_player == players[1] else players[1]

play_game()
```
